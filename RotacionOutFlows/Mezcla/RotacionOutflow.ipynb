```python
%pylab inline
from scipy.interpolate import interp1d
```

```python
wl=loadtxt("./wl.dat")
wl0=loadtxt("./wl0.dat")
```

```python
#wl0 es la distribucion plana inicial 
h=hist(wl0, bins=199)
intensidades_planas, lambdas_planas = h[0], h[1]
intensidad_plana_max = amax(intensidades_planas)
```

```python
#wl es como queda despues de los outflows
h2=hist(wl, bins=199)
```

```python
v_th=12.4
c=300000.0 #km/s
frecuencia_LyA_reposo=2.47*(10**15) #1/s

#CON:  TAU= 6   VMAX=100   A=0
espectro_simulado=loadtxt("./t6v100a0.dat")
velocidades_simuladas = espectro_simulado[:,0]*v_th #en km/s
intensidades_simuladas = espectro_simulado[:,1]

intensidad_simulada_max = amax(intensidades_simuladas)
intensidades_simuladas = intensidades_simuladas*intensidad_plana_max/intensidad_simulada_max

frecuencias_simuladas = frecuencia_LyA_reposo*(1 - velocidades_simuladas/c) #1/s
lambdas_simuladas = c*1000/frecuencias_simuladas #metros
lambdas_simuladas = lambdas_simuladas*10**10 #angstroms

plot(lambdas_simuladas,intensidades_simuladas)
```

```python
num_lambdas = size(lambdas_simuladas)
delta_lambda = lambdas_simuladas[2]- lambdas_simuladas[1]
lambda_simulada_inicial = lambdas_simuladas[0]
lambda_simulada_final = lambdas_simuladas[num_lambdas-1]
lambda_plana_inicial =  amin(lambdas_planas)
lambda_plana_final =  amax(lambdas_planas)
num_abajo = int(ceil((lambda_simulada_inicial- lambda_plana_inicial)/delta_lambda))
num_arriba = int(ceil((lambda_plana_final- lambda_simulada_final)/delta_lambda))

abajo = zeros(num_abajo)
arriba = zeros(num_arriba)

intensidades_simuladas_extendidas = concatenate((abajo,intensidades_simuladas,arriba))

for i in range(0,num_abajo):
    lambda_simulada_inicial -= delta_lambda
    abajo[num_abajo-i-1] = lambda_simulada_inicial
    
for i in range(0,num_arriba):
    lambda_simulada_final += delta_lambda
    arriba[i] = lambda_simulada_final

lambdas_simuladas_extendidas = concatenate((abajo,lambdas_simuladas,arriba))

plot(lambdas_simuladas_extendidas, intensidades_simuladas_extendidas)
```

```python
#interpolacion de simuladas
f_sim = interp1d(lambdas_simuladas_extendidas, intensidades_simuladas_extendidas)

intensidades_nuevas = f_sim(wl0)
```

```python
fig=figure()
espectro_final=hist(wl, bins=100, weights=intensidades_nuevas)
xlabel("$\lambda$ (Armstrongs)")
ylabel("Flux (Arbitrary Units)")
savefig("./espectrofinal.png",format = 'png')
show()
close()
```
